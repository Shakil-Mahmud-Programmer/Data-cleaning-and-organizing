```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Global Pwer Plants old.xlsx",header=None)
df = list(df[0])
```

```python
def count(string):
    val =0
    for i in string:
        val+=1
    return val
def check(string):
    try:
        float(string)
        return False
    except:
        return True

def popi(lis,val):
    for i in lis:
        if i == val:
            ind = lis.index(i)
            lis.pop(ind)
```

```python
data = []
flag =0 
for i in df:
    try:
        if i.isupper() and count(i) == 3 :
            data.append(flag)
    except:
        pass
    flag+=1
```

```python
for i in data:
    if check(df[i+1]) == False:
        popi(data,i)
```

```python
for i in data:
    try:
        if check(df[i+1]) == False:
            data.pop(df.index(i))
        else:
            continue
    except:
        continue
```

```python
outter = []
innner = []
for i in data:
    try:
        for j in range(i,data[data.index(i)+1]):
            innner.append(df[j])

        value = innner.copy()
        innner.clear()
        outter.append(value)
    except:
        continue
```

```python
df_new = pd.DataFrame(outter)
df_new.columns = ["country code","country","name of powerplant","capacity in MW","latitude","longitude","primary_fuel","secondary fuel","other_fuel 1"]
```
